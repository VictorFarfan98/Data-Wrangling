```python
import pandas as pd
import numpy as np
import re
from collections import Counter
import nltk
from textblob import TextBlob
nltk.download('wordnet')
nltk.download('punkt')
from nltk.stem import WordNetLemmatizer
```

# 2. Importar el dataframe generado en R a Python.

```python
data = pd.read_csv("tweets.csv")
data.head()
```

# 3. Utilizando expresiones regulares: 

## a. Crear una función normalizadora de texto que limpie los caracteres que pueden             generar ruido en el string del tweet y aplicarla a los mismos.

```python
def stringNormalizer(df):
    df = df.replace(to_replace='[^\sa-zA-Z0-9]@', value = "", regex=True)
    df = df.replace(to_replace='RT', value="", regex=True)
    return df

data = stringNormalizer(data)
data.head()
```

## b. Generar una nueva columna llamada "Handler" que muestre el handle del usuario.

```python
def addHandler(df):
    df.rename(columns={'screenName':'Handler'}, inplace=True)
    df['Handler'] = df['Handler'].replace(to_replace='[^\sa-zA-Z0-9]', value='', regex=True)
    
    return df
data = addHandler(data)
data.head()
```

## c.  Generar una nueva columna llamada "Source" que muestre de donde se generó            el tweet

```python
print(data)

def addSource(df):
    print(df)
    df['statusSource'] = df['statusSource'].replace('<.*?>', value='', regex=True)
    df.rename(columns={'statusSource':'Source'}, inplace=True)
    df['Handler'] = '@' + df['Handler'].astype(str)
    
    return df
data = addSource(data)    
data.head()
```

# 4.  Generar un corpus del dataset original que contenga: Handler, screenName, retweetCount y Source

```python
def corpusMaker(df):
    corpus = pd.DataFrame({'Handler':df['Handler'], 'retweetCount':df['retweetCount'], 
                         'Source':df['Source'], 'Text':df['text']})
    return corpus

corpus = corpusMaker(data)
corpus.head()
```

# 5. Generar una función que lematize cada tweet y guardar el resultado en una nueva columna llamada "lem_text" 

```python
lemma_array = []
lemmatizer = WordNetLemmatizer()
polarity = []
lem_polarity = []
subjectivity = []
lem_subjetivity = []

def lematizer(lemma, lemma_array):
    word_list = nltk.word_tokenize(lemma)
    lemmatized_output = ' '.join([lemmatizer.lemmatize(w) for w in word_list])
    lemma_array.append(lemmatized_output)    

for x in corpus['Text']:
    lematizer(x, lemma_array)
corpus['lem_text'] = lemma_array

corpus = corpus.sort_values(by=['Handler'])
corpus.head()
```

# 6.  Generar una función calcule la polaridad y subjetividad de las columnas "text" y "lem_text"

```python
def sentiment_analysis(text):
    polarity = []
    subjectivity = []
    for tw in text:
        tw = TextBlob(tw)
        ann = tw.sentiment
        polarity.append(ann[0])
        subjectivity.append(ann[1])

    return polarity,subjectivity
corpus['Polarity_Text'], corpus['Subjectivity_Text'] = sentiment_analysis(corpus.lem_text)
corpus['Polarity_LemText'], corpus['Subjectivity_LemText'] = sentiment_analysis(corpus.lem_text)

corpus = corpus.sort_values(by=['Polarity_Text'])
corpus.head()
```

# 7. Utilizando pandas, calcular la media de la polaridad y subjetividad de todos las columnas del inciso 6 del corpus. Comparar si la lematización afecta en el resultado

```python
rows = ['Polarity_Text', 'Subjectivity_Text', 'Polarity_LemText', 'Subjectivity_LemText']

for i in rows:
    print(i,":",corpus[i].mean())
```
